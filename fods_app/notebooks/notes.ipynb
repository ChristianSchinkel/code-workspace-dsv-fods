````python
"""This notebook contains notes and code snippets related to the Foundations of Data Science (FODS) course. It includes imports of necessary libraries, as well as a reference to the healthcare dataset used in Lab 0.
"""
import pandas as pd
# import fods_app.resources.stats


# Functions for loading the dataset
def root_dir() -> str:
    """Return the root directory for the dataset.

    Returns:
        str: The root directory path.
    """
    return "../data"


def data_path(filename: str) -> str:
    """Return the directory for a given filename.

    Args:
        filename (str): The name of the file.

    Returns:
        str: The directory path for the file.
    """
    return root_dir() + "/" + filename


def load_data(filename: str, chunk_size: int = -1) -> pd.DataFrame:
    """Load the healthcare dataset from a specified file path.

    Args:
        filename (str): The name of the file containing the dataset.
        chunk_size (int, optional): The number of rows in each chunk (default is -1).
    Returns:
        pd.DataFrame: A DataFrame containing the loaded dataset.
    """
    path = data_path(filename)
    _df = pd.read_csv(path)

    if chunk_size > 0:
        return choping(pd.read_csv(path), chunk_size=chunk_size)
    return _df


PATH = "healthcare_dataset.csv"

df = load_data(PATH)
df.head()

# Display Frequency of Features
df["Blood Type"].value_counts()

# Chopping data into smaller pieces
def choping(_df, chunk_size=1000):
    """Chop a DataFrame into smaller pieces.
    Args:
        _df (pd.DataFrame): The DataFrame to chop.
        chunk_size (int, optional): The number of rows in each chunk (default is 1000).
    Returns:
        list: A list of DataFrame chunks.
    """
    df_list = []
    for chunk in _df.groupby(_df.index // chunk_size):
        df_list.append(chunk[1])
    return df_list



# Grouping and Aggregating Data
def summarize(_df, group_by, value_col, agg="mean", sort=True):
    """Summarize a DataFrame by grouping and aggregating values.

    Args:
        _df (pd.DataFrame): The DataFrame to summarize.
        group_by (str): The column name to group by.
        value_col (str): The column name to aggregate.
        agg (str, optional): The aggregation function to use (default is "mean").
        sort (bool, optional): Whether to sort the results by the aggregated values (default is True).
    Returns:
        pd.DataFrame: A summarized DataFrame with the aggregated values.
    """
    results = _df.groupby(group_by)[value_col].agg(agg)
    if sort:
        results = results.sort_values(ascending=False)
    return results

# Identify Outliers in the Dataset
def above_percentile(_df, group_by, value_col, q=0.9) -> pd.DataFrame:
    """Identify rows in a DataFrame where the value in a specified column is above a certain percentile.

    Args:
        _df (pd.DataFrame): The DataFrame to analyze.
        group_by (str): The column name to group by.
        value_col (str): The column name to evaluate for outliers.
        q (float, optional): The percentile threshold to use for identifying outliers (default is 0.9).
    Returns:
        pd.DataFrame: A DataFrame containing rows where the value in the specified column is above the given percentile.
    """
    results = summarize(_df, group_by, value_col, agg="mean", sort=False)
    threshold = results.quantile(q)
    return _df[_df[value_col] > threshold]
````

### Q1 Data Representation and Dimentisionality Reduction

#### a. heterogenious types of attributes, such as numerical, categorical, and binary values
1. Challenges when computing similarity between data objects with mixed attribute types:
- Different attribute types (e.g., categorical vs. numerical) require different treatments. Applying the same
distance metric across all features can lead to meaningless results. For example, subtracting category labels like
”red” and ”blue” makes no sense. Each type must be encoded and scaled appropriately before computing similarity.
2. One approach is to encode all attributes into numeric vectors: for example, use one-hot encoding for categorical
variables, scale numerical features (e.g., using z-scores), and represent binary attributes as 0/1. The combined
vector can then be used with similarity metrics (e.g., cosine similarity) that handle mixed types more effectively.
#### b. Main differences between linear and non-linear dimensionality reduction methods
- Linear methods reduce dimensionality by projecting data onto a lower-dimensional linear subspace (e.g., PCA
uses linear combinations of features). Non-linear methods try to preserve more complex non-linear relationships
in the data, such as curved manifolds. An example of a linear method is PCA, which preserves directions of
maximum variance using the eigenvectors of the covariance matrix. A non-linear method like t-SNE preserves
local neighborhood relationships, which is useful when data lies on a non-linear manifold. Non-linear methods are
often preferred when the data has complex, non-linear structure that linear projections would fail to capture (e.g.,
visualizing clusters or curved patterns).
#### c. The "curse of dimensionality"
- The curse of dimensionality refers to the phenomenon where data becomes increasingly sparse as the number
of dimensions grows, making statistical estimation and learning harder. For example, in high dimensions, distance
metrics (like Euclidean distance) lose meaning — all points tend to become similarly distant. One mitigation
strategy is to apply dimensionality reduction, such as PCA or feature selection, to reduce the number of variables
while preserving the most informative structure.

1. How can PCA help reduce the dimensionality of this dataset?
PCA reduces dimensionality by finding new axes called principal components, that capture the most variance
in the data. These components are found by computing the eigenvectors of the covariance matrix. The data is then
projected onto the top components, which helps remove redundancy and keep only the important structure.

2. How would you decide how many principal components to keep, and what would you expect in this case?
If some features are linear combinations of others, the covariance matrix will have eigenvalues that are zero or
very small, showing those directions don’t add new information. We keep the number of components with large eigenvalues, which together explain most of the variance. In this case, A5 is perfectly correlated with A3 and thus redundant, while A1, A2 and A4 highly correlated with each other. Thus we would expect two principal components to be enough.

## Q2 Clustering
### a. Compare DBSCAN and K-Means clustering algorithms
DBSCAN groups points that are closely packed together based on density (using parameters ϵ and minPts) and identifies points in low-density regions as outliers. k-means partitions data into k clusters by minimizing within-cluster variance, assuming roughly spherical and equally sized clusters. Two strengths of DBSCAN are that it can discover clusters of arbitrary shape, and that it automatically detects outliers. Two limitations of DBSCAN are that it requires tuning two parameters (ϵ, minPts), and it doesn’t perform well when densities vary across clusters.

### b. Silhouette score
**You run two clustering algorithms on the same dataset. One gives a high silhouette score, the other gives high purity (with respect to known class labels).**
1. What do these two metrics measure?
- Silhouette score is an intrinsic evaluation metric that measures how similar an object is to its own cluster
compared to other clusters. It captures cohesion and separation. Purity is an extrinsic metric that compares
clusters to ground truth class labels. It reflects how well clusters align with known labels.
2. Can one clustering be better in terms of silhouette score, but worse in terms of purity? Explain why this might happen.
- Yes, this can happen. A clustering may have high silhouette score if it finds well-separated, compact clusters — but those clusters might not match the true labels, resulting in low purity. Conversely, a clustering that aligns with labels might be less compact, lowering its silhouette score.

Single Linkage Distance: $d(\{A\}, \{B,C\}) = \min_{x \in A,\; y \in \{B,C\}} d(x,y)$

Complete Linkage Distance: $d(\{A\}, \{B,C\}) = \max_{x \in A,\; y \in \{B,C\}} d(x,y)$

##### Distances betwwen points:
- $d(A,B) = 4$
- $d(A,C) = 6$

Single Linkage Distance: $d(\{A\}, \{B,C\}) = \min(4, 6) = 4$

Complete Linkage Distance: $d(\{A\}, \{B,C\}) = \max(4, 6) = 6$

Thus, the single linkage distance between clusters {A} and {B,C} is 4, while the complete linkage distance is 6.

### b. Naïve Bayes Classifier
A Naïve Bayes classifier is trained to predict whether an email is Spam or Not Spam based on three features:
Feature: Presence of promotional keywords: “free”, “win”.

1. Naive bayes is based on bayes rule which can be stated as follows:
Given a prior, on collecting some evidence (likelihood) we update the prior to result in posterior formulated as follows:
$P(Y|X) = \frac{P(X|Y)P(Y)}{P(X)}$

Where $Y$ is the class label (Spam or Not Spam) and $X$ is the feature vector (presence of keywords).
2. It is called naive bayes, as features are assumed to be independent of each other which is a naive assumption. Based on above example words are assumed to be independent of each other when computing likelihood.
$P(X|Y) = P(X_1|Y)P(X_2|Y)P(X_3|Y)$

Where $X_1$, $X_2$ and $X_3$ are the features representing presence of keywords.
